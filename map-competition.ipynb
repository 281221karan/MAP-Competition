```python
model_paths = [
               ["/kaggle/input/delhermes-2600/Qwen3-0.6B/checkpoint-8000","/kaggle/input/delhermes-2600/Qwen3-0.6B/checkpoint-8000",0.5],
               ["/kaggle/input/qwen2-5-0-5b-instruct/Qwen2.5-0.5B-Instruct","/kaggle/input/qwen-2000/Qwen2.5-0.5B-Instruct/checkpoint-8000",0.3],
               ["/kaggle/input/delphermes/Qwen3-0.6B/checkpoint-8000","/kaggle/input/delphermes/Qwen3-0.6B/checkpoint-8000",0.2]
              ]
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification, AutoModelForCausalLM
from tqdm.auto import tqdm,trange
import torch.nn as nn
from datasets import Dataset
from sklearn.metrics import f1_score, precision_recall_fscore_support, accuracy_score
from transformers import DataCollatorWithPadding
from transformers import TrainingArguments
from transformers import Trainer
import gc
import torch

def clear_memory():
    gc.collect()
    torch.cuda.empty_cache()
    
all_logits = None
weighted_logits = None

for i in model_paths:
    print(i[0],i[1])
    
    tokenizer = AutoTokenizer.from_pretrained(i[0])
    model = AutoModelForSequenceClassification.from_pretrained(
                                                                i[1], 
                                                                num_labels = len(le.classes_),
                                                                device_map = "auto"
                                                                )
    tokenizer.pad_token = tokenizer.eos_token
    tokenizer.padding_side = "right" 
    model.config.pad_token_id = tokenizer.pad_token_id
    
    train_dataset = Dataset.from_pandas(train_df[["input_text","labels"]])
    test_dataset = Dataset.from_pandas(test_df["input_text"].to_frame(name = "input_text"))
    train_dataset,test_dataset
    
    train_dataset[0],test_dataset[0]
    
    split = train_dataset.train_test_split(test_size = 0.2, seed = 42)
    train_dataset = split["train"]
    val_dataset = split["test"]
    train_dataset,val_dataset
    
    def tokenize(examples):
        return tokenizer(examples["input_text"], padding="max_length", truncation=True,max_length = 256)
    
    train_dataset = train_dataset.map(tokenize, batched = True)
    val_dataset = val_dataset.map(tokenize, batched = True)
    test_dataset = test_dataset.map(tokenize, batched = True)
    train_dataset,val_dataset,test_dataset
    
    cols = ["input_ids", "attention_mask", "labels"]
    train_dataset.set_format(type="torch", columns = cols)
    val_dataset.set_format(type="torch", columns = cols)
    test_dataset.set_format(type="torch", columns = ["input_ids", "attention_mask"])

    def compute_metrics(p):
        predictions = np.argmax(p.predictions, axis=1)
        labels = p.label_ids
        f1 = f1_score(labels, predictions, average='weighted')
    
        accuracy = accuracy_score(labels, predictions)
        precision, recall, _, _ = precision_recall_fscore_support(labels, predictions, average='weighted')
    
        return {
            "f1": f1,
            "accuracy": accuracy,
            "precision": precision,
            "recall": recall
        }
    
    data_collator = DataCollatorWithPadding(tokenizer=tokenizer, pad_to_multiple_of=8)
    if i[0] == "/kaggle/input/delhermes-2600/Qwen3-0.6B/checkpoint-8000":
        train_args = TrainingArguments(
            # output_dir=f"./{model_name.split('/')[1]}",
            max_steps = 8000,
            per_device_train_batch_size=8,
            per_device_eval_batch_size=16,
            eval_strategy="steps",
            logging_strategy="steps",
            save_strategy="steps",
            save_steps = 200,
            eval_steps = 200,
            weight_decay=0.01,
            logging_steps = 200,
            learning_rate=3e-5,
            save_total_limit = 1,
            report_to="none",
        )
    elif i[0] == "/kaggle/input/qwen2-5-0-5b-instruct/Qwen2.5-0.5B-Instruct":
        train_args = TrainingArguments(
        # output_dir=f"./{model_name.split('/')[1]}",
        max_steps = 4000,
        per_device_train_batch_size=8,
        per_device_eval_batch_size=16,
        eval_strategy="steps",
        logging_strategy="steps",
        save_strategy="steps",
        save_steps = 200,
        eval_steps = 200,
        weight_decay=0.01,
        logging_steps = 200,
        learning_rate=2e-5,
        save_total_limit = 2,
        report_to="none",
    )
    else:
        train_args = TrainingArguments(
            # output_dir=f"./{model_name.split('/')[1]}",
            max_steps = 8000,
            per_device_train_batch_size=8,
            per_device_eval_batch_size=16,
            eval_strategy="steps",
            logging_strategy="steps",
            save_strategy="steps",
            save_steps = 200,
            eval_steps = 200,
            weight_decay=0.01,
            logging_steps = 200,
            learning_rate=2e-5,
            save_total_limit = 2,
            save_only_model=True,
            report_to="none",
        )
            
    trainer = Trainer(
        model=model,
        args=train_args,
        train_dataset=train_dataset,
        eval_dataset=val_dataset,
        data_collator=data_collator,
        compute_metrics=compute_metrics
    )
    # trainer.train()
    test_dataset = test_dataset.map(tokenize, batched = True)
    prediction = trainer.predict(test_dataset)
    logits = prediction.predictions

    weighted_model_logits = logits * i[2]
    if weighted_logits is None:
        weighted_logits = weighted_model_logits
    else:
        weighted_logits += weighted_model_logits
    
    # if all_logits is None:
    #     all_logits = np.zeros_like(logits)
    # all_logits += logits
    
    del tokenizer, model, trainer
    clear_memory()

# ensemble_logits = all_logits / len(model_paths)

top3_indices = np.argsort(weighted_logits, axis=1)[:, -3:][:, ::-1]
top3_labels = np.array([le.inverse_transform(row) for row in top3_indices])
top3_combined = [" ".join(row) for row in top3_labels]

# Create submission
submission = pd.DataFrame({
    "row_id": test_df.row_id.values,
    "Category:Misconception": top3_combined,
})

submission.to_csv('submission.csv', index=False)

# top3_indices = np.argsort(ensemble_logits, axis=1)[:, -3:][:, ::-1]
# top3_labels = np.array([le.inverse_transform(row) for row in top3_indices])
# top3_combined = [" ".join(row) for row in top3_labels]

# submission = pd.DataFrame({
#     "row_id": test_df.row_id.values,
#     "Category:Misconception": top3_combined,
# })

# submission.to_csv('submission.csv', index=False)
```
